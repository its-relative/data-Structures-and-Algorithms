```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
pd.set_option("display.max_rows",10)
pd.set_option("display.max_columns", 1440)
```

```python
df = pd.read_csv("ToyotaCorolla.csv",encoding="ISO-8859-1")
df
```

```python
y = df.iloc[:,2:3]
y
x = df[["Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"]]
x
```

```python
new_x = pd.get_dummies(x)
new_x
```

```python
sc = StandardScaler()
new_x1 = sc.fit_transform(new_x)
# new_x1
```

```python
new_x1.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_x1,y,test_size=0.2,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
comp = pd.DataFrame({"pred":np.array(y_pred).reshape(-1),"actuals":np.array(y_test).reshape(-1)})
```

```python
comp
```

```python
r2_score(y_true=y_test,y_pred=y_pred)
```
