```python
import pandas as pd
import numpy as np
```

```python
data  = pd.read_csv("Company_Data.csv")
df  = data.copy()
```

```python
df
```

```python
df = pd.get_dummies(df).astype(float)
```

```python
df
```

```python
y = df.iloc[:,:1]
y
```

```python
x = df.iloc[:,1:]
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeRegressor

deciReg = DecisionTreeRegressor(random_state=42)
deciReg.fit(x_train,y_train)
```

```python
y_pred = deciReg.predict(x_test)
y_pred
```

```python
check_df = pd.DataFrame({"pred":y_pred,"actual":np.array(y_test).reshape(-1)})
```

```python
check_df
```

```python
from sklearn.metrics import accuracy_score,r2_score

r2_score(y_test,y_pred)
```
