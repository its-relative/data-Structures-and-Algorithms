```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("bank-full.csv",sep = ";")
df = data.copy()
```

```python
df
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1:]
```

```python
x
```

```python
y = y.replace("no",0)
```

```python
y = y.replace("yes",1)
```

```python
y
```

```python
check = x.copy()
```

```python
check
```

```python
pd.set_option("display.max_columns",60)
```

```python
check = pd.get_dummies(check).astype(int)
check
```

```python
from sklearn.preprocessing import StandardScaler
stdscale = StandardScaler()
check = stdscale.fit_transform(check)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(check,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
last = pd.DataFrame({"pred":np.array(y_pred).reshape(-1),"actual":np.array(y_test).reshape(-1)})
last
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
confusion_matrix(last["actual"],last["pred"])
```

```python
precision_score(last["actual"],last["pred"])
```

```python
recall_score(last["actual"],last["pred"])
```

```python
accuracy_score(last["actual"],last["pred"])
```
