```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Fraud_check.csv")
df = data.copy()
```

```python
new_risk = df[df["Taxable.Income"] <= 30000]
new_good = df[df["Taxable.Income"] > 30000]
```

```python
new_risk["Risky"] = 1
```

```python
new_good["Risky"] = 0
```

```python
df = pd.concat([new_good,new_risk], axis= 0)
```

```python
df
```

```python
# df.iloc[:,:-1] = pd.get_dummies(df.iloc[:,:-1]).astype(int)
```

```python
mid_df = df.drop("Risky",axis=1) 
```

```python
# df = pd.get_dummies(df).astype(int)
```

```python
mid_df = pd.get_dummies(mid_df).astype(int)
```

```python
mid_df
```

```python

mid_df["Risky"] = df["Risky"]
```

```python
mid_df
```

```python
y = mid_df.iloc[:,-1:]
y
```

```python
x = mid_df.drop("Risky",axis=1)
x
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dc = DecisionTreeClassifier(random_state=42,max_depth=7)
dc.fit(x_train,y_train)


```

```python
y_pred = dc.predict(x_test)
```

```python
check = pd.DataFrame({"pred":y_pred,"actual":np.array(y_test).reshape(-1)})
```

```python
check
```

```python
from sklearn.metrics import recall_score,precision_score,accuracy_score
```

```python
recall_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```
